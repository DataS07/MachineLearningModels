### Get Features and class labels into a NumPy Arrays

```python
X = df[['RM']].values
y = df['MEDV'].values
```

## LinearRegression model

```python
class LinearRegressionGD():
    def __init__(self, X, y, alpha=0.003, n_iter=1500):
        self.alpha = alpha
        self.n_iter = n_iter
        self.n_samples = len(y)
        self.n_features = np.size(X, 1)
        self.X = np.hstack((np.ones(
            (self.n_samples, 1)), (X - np.mean(X, 0)) / np.std(X, 0)))
        self.y = y[:, np.newaxis]
        self.params = np.zeros((self.n_features + 1, 1))
        self.coef_ = None
        self.intercept_ = None

    def fit(self):
        self.cost_ = []
        for i in range(self.n_iter):
            self.params = self.params - (self.alpha/self.n_samples) * \
            self.X.T @ (self.X @ self.params - self.y)
            self.cost_.append(self.computeCost(X,y))
        self.intercept_ = self.params[0]
        #print(self.intercept_)
        self.coef_ = self.params[1:]
        return self

    def computeCost(self, X=None, y=None):
        if X is None:
            X = self.X
        else:
            n_samples = np.size(X, 0)
            X = np.hstack((np.ones(
                (n_samples, 1)), (X - np.mean(X, 0)) / np.std(X, 0)))
        if y is None:
            y = self.y
        else:
            y = y[:, np.newaxis]
        y_pred = X @ self.params
        cost = (((y - y_pred)**2).sum() / ((y - y.mean())**2).sum())
        return cost
    
    def predict(self, X):
        n_samples = np.size(X, 0)
        y = np.hstack((np.ones((n_samples, 1)), (X-np.mean(X, 0)) \
                            / np.std(X, 0))) @ self.params
        return y
    def get_params(self):
        return self.params
```

```python

def lin_regplot(X,y,model):
    plt.scatter(X,y,c='steelblue',edgecolor='white',s=70)
    plt.plot(X,model.predict(X),color='black',lw=2)
    return

slr = LinearRegressionGD(X,y,alpha=0.1,n_iter=150)
slr.fit()

y_pred = slr.predict(X)
print('Slope: %.3f' % slr.coef_)
print('Intercept: %.3f' % slr.intercept_)
lin_regplot(X,y,slr)
plt.xlabel('Average number of rooms [RM]')
plt.ylabel('Price in $1000s [MEDV]')
plt.show()
```

```python
plt.plot(range(len(slr.cost_)), slr.cost_, 'r')
plt.xlabel('Nb of iterations')
plt.ylabel('Errors (Cost Function)')
plt.show()
```

## Split and Standardization

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state=0)
RM=np.array(X_train[:,0])
MEDV=np.array(y_train)

sc_x = StandardScaler()
sc_y = StandardScaler()
X_train_std = sc_x.fit_transform(X_train)
y_train_std = sc_y.fit_transform(y_train[:,np.newaxis]).flatten()
X_test_std = sc_x.fit_transform(X_test)
y_test_std = sc_y.fit_transform(y_test[:,np.newaxis]).flatten()

slr1 = LinearRegressionGD(X_train_std,y_train_std,alpha=0.1,n_iter=150)
slr1.fit()

y_pred_train_std = slr1.predict(X_train_std)
y_pred_test_std = slr1.predict(X_test_std)
print('Slope: %.3f' % slr1.coef_)
print('Intercept: %.3f' % slr1.intercept_)
lin_regplot(X_train_std,y_train_std,slr1)
plt.xlabel('Average number of rooms [RM]')
plt.ylabel('Price in $1000s [MEDV]')
plt.show()
```

```python
plt.plot(range(len(slr1.cost_)), slr1.cost_, 'r')
plt.xlabel('Nb of iterations')
plt.ylabel('Errors (Cost Function)')
plt.show()
```

####               We need around 40 epochs for convergence (with keeping the same learning Rate)

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

print('MSE train: %.3f, test: %.3f' % (mean_squared_error(y_train_std,y_pred_train_std),mean_squared_error(y_test_std,y_pred_test_std)))
print('R^2 train: %.3f, test: %.3f' % (r2_score(y_train_std,y_pred_train_std),r2_score(y_test_std,y_pred_test_std)))
```

```python
from sklearn.linear_model import LinearRegression
sklearn_reg = LinearRegression().fit(X_train_std, y_train_std)

skl_train_acc = sklearn_reg.score(X_train_std, y_train_std)
c=slr1.computeCost()
acc=1-c
skl_test_acc = sklearn_reg.score(X_test_std, y_test_std)
ct=slr1.computeCost(X_test_std, y_test_std)
acct=1-ct
print('train_Accuracy: %.3f' % acc)
print('sklearn_train Accuracy: %.3f' % skl_train_acc)
print('---------------------------------------------')
print('test_Accuracy: %.3f' % acct)
print('sklearn_test Accuracy: %.3f' % skl_test_acc)
```
